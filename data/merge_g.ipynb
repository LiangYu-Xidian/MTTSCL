```python
import pandas as pd
import numpy as np
```

```python
g_neg = pd.read_csv("./g_neg.csv")
```

```python
g_neg.rename(columns={'label':'label_g_neg'},inplace=True)
g_neg
```

```python
g_pos = pd.read_csv("./g_pos.csv")
g_pos.rename(columns={'label':'label_g_pos'},inplace=True)
g_pos
```

```python
g_neg_pos_df = pd.concat([g_neg,g_pos])
g_neg_pos_df
```

```python
g_neg_pos_df = g_neg_pos_df.fillna("no_label")
```

```python
g_neg_pos_df.to_csv("./g_neg_pos.csv",index = False)
```
